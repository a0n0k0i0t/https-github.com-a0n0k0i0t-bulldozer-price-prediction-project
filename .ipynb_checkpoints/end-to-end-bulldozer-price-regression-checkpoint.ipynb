# Prediction the Sale Price of Bulldozers using Machine Learning

Predict the auction sale price for a piece of heavy equipment to create a "blue book" for bulldozers.

## 1.Problem
The goal is to predict the sale price of a particular piece of heavy equiment at auction based on it's usage, equipment type, and configuaration.  The data is sourced from auction result postings and includes information on usage and equipment configurations.

## 2.Data
The data is split into three parts:

<ul>
    <li>Train.csv is the training set, which contains data through the end of 2011.
    <li>Valid.csv is the validation set, which contains data from January 1, 2012 - April 30, 2012 
    <li>Test.csv is the test set, which won't be released until the last week of the competition. It contains data from  May 1, 2012 - November 2012.
</ul>

## 3.Evaluation
The evaluation metric for this competition is the RMSLE (root mean squared log error) between the actual and predicted auction prices.
For more information:
https://www.kaggle.com/c/bluebook-for-bulldozers/overview/evaluation

## 4.Features

### Converting `string` to `catergory`
Pandas has a datatype `category` which treats string values as integers.
Since we can't train a model with only string values we need to convert it to category

```python
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        df[label] = content.astype("category").cat.as_ordered()
```

### Handling missing values in the table
Now that we have suitable data types the only thing left in preprocessing is to handle missing data

```python
#Checking percentage of missing data
df.isna().sum()/len(df)
```

### Filling missing value
#### 1. Fill missing numeric values

```python
#Find out which labels are numeric
numeric_labels_missing=[]
for feature,d_type in df.items():
    if pd.api.types.is_numeric_dtype(d_type) and df[feature].isna().sum():
        numeric_labels_missing.append(feature)

numeric_labels_missing
```

```python
# Filling missing value with median.
# There are different strategies to fill missing values like fill by mean,mode,median
# But median is more robust and works well even if data is not well distributed
for label in numeric_labels_missing:
    df[label+"_is_missing"]=pd.isnull(label)
    df[label]=df[label].fillna(df[label].median())
```

```python
# Checking if we got rid of missing values in numeric labels
for label in numeric_labels_missing:
    print(label,df[label].isna().sum())
```

#### 2. Fill missing categorical values

```python
#Finding categorical labels
categoric_labels=[]
for feature,d_type in df.items():
    if not pd.api.types.is_numeric_dtype(d_type):
        categoric_labels.append(feature)

categoric_labels
```

```python
# Turning categoricals to numerics and filling missing data
for label in categoric_labels:
    # For category variables 'Nan' is coded as -1.
    # Incrementing Categorical(label).code makes 'Nan' code as 0
    df[label+"_is_missing"]=pd.isnull(label)
    df[label]=pd.Categorical(df[label]).codes+1
```

```python
# Checking if we got rid of missing values in categoric labels
for label in categoric_labels:
    print(label,df[label].isna().sum())
```

# Building model
This is the main part for prediction of the SalePrice. The performance of the predcition is highly dependent on kind of model we use and how well we train the model

### Splitting data for train and validation
One can try and train the whole data and then validate the same data. But there's a problem with that. The model which uses whole data to train can predict the same data with almost 99%-100% accuracy but might work terribly for some new data. That's why using different train set and validation set is used.

```python
#Splitting data
from sklearn.model_selection import train_test_split
X=df.drop("SalePrice",axis=1)
y=df.SalePrice
X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.3)
```

### Building evaluation function
We will use the same evaluation function as used in competition **Root Mean Squared Log Error (RMSLE)**

```python
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, r2_score

#Using different error metrics to evaluate the error of a model

# Function for RMSLE using mean_squared_log_error
def rmsle(y_test,y_pred):
    return np.sqrt(mean_squared_log_error(y_test,y_pred))

def model_score(model):
    train_pred=model.predict(X_train)
    val_pred=model.predict(X_valid)
    scores={
        "Training MAE" : mean_absolute_error(y_train,train_pred),
        "Valid MAE" : mean_absolute_error(y_valid,val_pred),
        "Training RMSLE" : rmsle(y_train,train_pred),
        "Valid RMSLE" : rmsle(y_valid,val_pred),
        "Training Score" : r2_score(y_train,train_pred),
        "Valid Score" : r2_score(y_valid,val_pred)
    }
    return scores
```

### Using RandomForestRegressor as the model
RandomForestRegressor is a trusted model for regression.We will try it out

```python
len(X_train)
```

```python
# X_train has large size. Training such a data will have a high training time.
# We will use reduced data for improving training time
# Using only 10000 sample (a small chunk of training data) reduces training time significantly
print(10000/len(X_train)*100,'%')
```

```python
from sklearn.ensemble import RandomForestRegressor
# We will be only using 10000
model= RandomForestRegressor(n_jobs=-1,random_state=42,max_samples=10000)
```

```python
%%time
#Fitting the model
model.fit(X_train,y_train)
```

Now that the model has been trained lets evaluate the model.

```python
model_score(model)
```

## Tuning Hyperparameters of RandomForestRegressor
Tuning hyperparameters will find the such values of parameters which makes the best model

```python
%%time
from sklearn.model_selection import RandomizedSearchCV

# Trying out different values of RandomForestRegressor hyperparameters

rf_params={"n_estimators" : np.arange(10,100,10),
           "max_depth": [None,3,5,10],
           "min_samples_split": np.arange(2,20,2),
           "max_features": [0.1,0.5,1,"sqrt","auto"],
           "max_samples": [10000]
          }
#RandomizedSearchCV for tuning
rscv=RandomizedSearchCV(RandomForestRegressor(n_jobs=-1,random_state=42),
                       param_distributions=rf_params,
                        n_iter=20,
                        cv=5,
                        verbose=True
                       )
#Fitting RandomizedSearchCV model for hyperparameter tuning
rscv.fit(X_train,y_train)
```

This the best parameters found for RandomForestRegressor

```python
rscv.best_params_
```

```python
model_score(rscv)
```

### Training whole data on the best parameters

```python
%%time
best_model=RandomForestRegressor(n_estimators=30,
                                min_samples_split=4,
                                 max_samples=None,
                                max_features=0.5,
                                 max_depth=None,
                                 n_jobs=-1
                                )
best_model.fit(X,y)
```

```python
#Scores of best model
model_score(best_model)
```

### Preprocessing the test data
We might want to run predictions on test data but our test data is not in same format as of train data.
To bring the test data in format of train data we need to do the same preprocessing on test data as done in train data to bring same format like train data

```python
# We will run all the same preprocessing done for train data in test data
def preprocess_data(df):
    df['saledate']=df['saledate'].str.replace(" 0:00","")
    df['saledate']=pd.to_datetime(df['saledate'])
    df.sort_values(by=['saledate'],inplace=True,ascending=True)
    df_copy=df.copy()
    df["saleYear"]=df.saledate.dt.year
    df["saleMonth"]=df.saledate.dt.month
    df["saleDay"]=df.saledate.dt.day
    df["saleDayofWeek"]=df.saledate.dt.dayofweek
    df["saleDayofYear"]=df.saledate.dt.dayofyear
    df.drop("saledate",axis=1,inplace=True)
    for x in df.columns:
        if(df[x].dtype=='object'):
            df[x]=df[x].astype('category')
    numeric_labels_missing=[]
    for feature,d_type in df.items():
        if pd.api.types.is_numeric_dtype(d_type) and df[feature].isna().sum():
            numeric_labels_missing.append(feature)
            
    for label in numeric_labels_missing:
        df[label+"_is_missing"]=pd.isnull(label)
        df[label]=df[label].fillna(df[label].median())
        
    categoric_labels=[]
    for feature,d_type in df.items():
        if not pd.api.types.is_numeric_dtype(d_type):
            categoric_labels.append(feature)

    categoric_labels
    for label in categoric_labels:
        df[label+"_is_missing"]=pd.isnull(label)
        df[label]=pd.Categorical(df[label]).codes+1
```

This numbers don't make much sense lets see some visuals to have better understanding

```python
import seaborn as sns

def plot_important_features(columns,importance,n=25):
    plot_data=(pd.DataFrame({
        'features': columns,
        'feature_importance': importance})
        .sort_values('feature_importance',ascending=False).reset_index(drop=True))
    sns.barplot(x='feature_importance',
                y='features',
                data=plot_data[:n],
                orient='h')
```
